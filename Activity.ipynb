```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('tips.csv')
avg_tip_day = df.groupby('day')['tip'].mean().reset_index()

sns.barplot(x='day', y='tip', data=avg_tip_day)
plt.title('Average Tip by Day')
plt.show()
```

```python
total_bill_day_time = df.groupby(['day', 'time'])['total_bill'].sum().unstack()

total_bill_day_time.plot(kind='bar', stacked=True)
plt.title('Total Bill by Day and Time')
plt.show()
```

```python
avg_data = df.groupby('day')[['total_bill', 'tip']].mean()

avg_data.plot()
plt.title('Average Total Bill and Tip Over Days of the Week')
plt.ylabel('Amount ($)')
plt.show()
```

```python

df['total_bill'].plot(kind='hist', bins=20)
plt.title('Distribution of Total Bill')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=df)
plt.title('Total Bill vs Tip by Gender')
plt.show()
```

```python
# Select specific columns and calculate correlation
corr = df[['total_bill', 'tip', 'size']].corr()

# Plot the heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', hue='sex', data=df)
plt.title('Total Bill by Day and Gender')
plt.show()
```
